```python
# tách cột SaleDate thành 3 cột ngày_tháng_năm
def split_date (df, col):
    ##Step 1. Create array
    day = []
    month = []
    year = []
    #step 2. Scan all data 
    for i in range (fdata.shape[0]):
        d = df[col][i].date().day
        m = df[col][i].date().month
        y = df[col][i].date().year
        day.append(d)
        month.append(m)
        year.append(y)
        i = i+1
    #Step 3. Create new column after add array in the column
    df['day'] = day
    df['month'] = month
    df['year'] = year
    
split_date(fdata, 'SaleDate')     
    
```

# Đánh giá mức độ hài lòng của người tiêu dùng cho các cửa hàng

```python
# Each attribute we'll plot in the radar chart.
labels = ['MucDoHaiLong', 'YeuToThuongHieuSP', 'YeuToUyTinNhaBanle', 'YeuToBaoHanh', 'YeuToKhuyenMai']

values = point.loc[0].tolist()

# Number of variables we're plotting.
num_vars = len(labels)

# Split the circle into even parts and save the angles
# so we know where to put each axis.
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to "complete the loop"
# and append the start value to the end.
values += values[:1]
angles += angles[:1]

# ax = plt.subplot(polar=True)
fig, ax = plt.subplots(figsize=(17,10),subplot_kw=dict(polar=True))

# Draw the outline of our data.
# ax.plot(angles, values, color='red', linewidth=1)
# # Fill it in.
# ax.fill(angles, values, color='red', alpha=0.25)

# ax.set_rgrids([1, 2, 3, 4, 5])
# ax.set_thetagrids([0,72,144,216,288], labels) #do mặc định là chia 8 phần nhưng chỉ có 5 cột nên phải set lại phần góc
# ax.set_theta_offset(np.pi / 2)
# ax.set_theta_direction(-1)

#ax.set_thetagrids(np.degrees(angles), labels)

def add_to_radar(indexshop, color):
    values = point.loc[indexshop].tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=1, label=labels)
    ax.fill(angles, values, color=color, alpha=0.25)


#Add each car to the chart.
add_to_radar(0, 'red')
add_to_radar(1, '#00FFFF')
add_to_radar(3, '#429bf4')
add_to_radar(4, '#d42cea')
add_to_radar(5, 'black')


#ax.set_rgrids([1, 2, 3, 4, 5])
ax.set_rgrids([0.625, 1.25, 1.875, 2.5, 3.125, 3.75, 4.375, 5])

ax.set_thetagrids([0,72,144,216,288], labels) #do mặc định là chia 8 phần nhưng chỉ có 5 cột nên phải set lại phần góc
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

ax.tick_params(colors='#222222')
# Make the y-axis (0-100) labels smaller.
ax.tick_params(axis='y', labelsize=9)
# Change the color of the circular gridlines.
ax.grid(color='#AAAAAA')
# Change the color of the outermost gridline (the spine).
ax.spines['polar'].set_color('#222222')
# Change the background color inside the circle itself.
ax.set_facecolor('#FAFAFA')

# Add title.
ax.set_title('Đánh giá mức độ hài lòng của khách hàng từng shop', y=1.08)

# Add a legend as well.
ax.legend(labels=NameShop,loc='upper right', bbox_to_anchor=(1.6,1.1))
```
